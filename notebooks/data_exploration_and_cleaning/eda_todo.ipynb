```python
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Memuat variabel lingkungan dari file .env
load_dotenv()

# Mengakses variabel lingkungan yang diatur dalam file .env
project_path = os.getenv("PROJECT_FOLDER")
dataset_folder = 'data/raw_data/trip_advisor'
```

```python
# List of dataset files
files = [
    f"{project_path}/{dataset_folder}/json/balikpapan_todo.json",
    f"{project_path}/{dataset_folder}/json/berau_todo.json",
    f"{project_path}/{dataset_folder}/json/bontang_todo.json",
    f"{project_path}/{dataset_folder}/json/kutai_todo.json",
    f"{project_path}/{dataset_folder}/json/paser_todo.json",
    f"{project_path}/{dataset_folder}/json/penajam_todo.json",
    f"{project_path}/{dataset_folder}/json/samarinda_todo.json",
]

# Read and concatenate all datasets into a single DataFrame
dfs = []
for file in files:
    with open(file, encoding='utf-8') as json_file:
        data = json.load(json_file)
        df = pd.DataFrame(data)
        dfs.append(df)

df = pd.concat(dfs)
```

```python
# place_name, city, address, min_price, max_price, rating, latitude, longitude, opening_hours, contact_number, website
# activity, category,description
df['city'] = df['address_obj'].apply(lambda x: x['city'])
new_df = df[['location_id','name','city','rating','latitude','longitude','hours','phone','photo','address']].rename(
    columns={'name': 'place_name','phone':'contact_number','hours':'opening_hours'})
```

```python
new_df['price'] = ''
for i in range(len(new_df)):
    price = input(f"Masukkan harga untuk baris {i+1}: ")
    new_df.at[i, 'price'] = price
```
